```python
#Need: pre set working directory as where the data is

def parse_grade(file_name):
    
    #create a list to store output
    output = []

    #open data file
    with open("data1.yml", 'r') as f:
    
        #create variables:  lecture name, assignment name, max score and student who received the max score
        lec = ''
        assignment = ''
        max_score = 0
        max_student = ''
    
    
        #loop through each line in file
        for line in f:
        
            #if starts with lecture, update lecture name in variable lec (remove spaces)
            if line.strip().startswith('lecture'):
                lec = line.strip().replace(':', '')
            
       
    
    
            #if starts with assignment (remove spaces):
        
            # step 1: the last assignment's records end 
            #--> append the last assignment's max score and student name in output
        
            # step 2: the start of a new assignmnet
            #--> updata assignment name, and re_initialize score and student variable
            if line.strip().startswith('assignment'):
            
                A_id = file_name +'/' + lec + '/' + assignment
                output.append([A_id, max_student])
            
            
                assignment = line.strip().replace(':', '')
                max_score = 0
                max_student = ''
        
        
            #if ends with a digit
            #extract score in the line: 
            #if > max score, update max score and student name, else continue
            if line.strip() != '' and line.strip()[-1].isdigit():
                break_line = line.split(':')
            
                score =  float(break_line[-1].strip())
            
                if score > max_score:
                    max_score = score
                    max_student = break_line[0].strip()
                else:
                    continue
            else:
                continue
    
        #wrap up the last assignment's record
        A_id = file_name +'/' + lec + '/' + assignment
        output.append([A_id, max_student])
        
    #convert output list to dataframe (get rid of the first line)
    #-->first line is empty data
    # concate with example output
    df_output = pd.DataFrame(output[1:], columns=['assignment_id', 'highest_grade'])
    
    #save output as csv file
    df_output.to_csv('output.csv')
    
    #return the output dataframe (optional, if output is very large)
    return df_output

            
              
    
```

```python
#run program on our data 
parse_grade('data1.yml')
```
